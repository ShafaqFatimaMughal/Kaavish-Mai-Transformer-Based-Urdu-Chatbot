```python
#Imports
import os
import pandas as pd
import numpy as np
import os
from google.colab import files
import re

# Flag for dataset language
RomanUrdu = 0
```

```python
!git clone https://github.com/ShafaqFatimaMughal/Kaavish-Mai-Transformer-Based-Urdu-Chatbot.git
path = "/content/Kaavish-Mai-Transformer-Based-Urdu-Chatbot/Datasets/Initial/Self-Curated"
```

# Extracting Data

```python
## If we want to add datasets to the runtime memory, we can create a folder "Self-Curated"
## And add all files to that. Then use self_path instead of path
# self_path = "./Self-Curated"

data = {}

for root, dirs, filess in os.walk(path): # We need the files in this path / diretory

    for index, name in enumerate(filess): 
        naming_index = name.split("/")[-1].split(" ")[1] # We want just the filename
        filename = os.path.join(root, name)
        print(filename)

        if RomanUrdu:
          df = pd.read_csv(filename, usecols = ['Intent (Roman Urdu)', 'Response (Roman Urdu)'])
        else:
          df = pd.read_csv(filename, usecols = ['Intent', 'Response'])
        # df = df[df['Relevance'] != 0]
        # print(df.head())
        data[naming_index] = df # Save df in dict
        
if RomanUrdu:
  dataset = pd.DataFrame(columns=['Intent (Roman Urdu)', 'Response (Roman Urdu)'])
else:
  dataset = pd.DataFrame(columns=['Intent', 'Response'])

# Concatenating dataset together
for i in data.keys(): 
  dataset = dataset.append(data[i])
dataset.info
```

# Cleaning Data

```python
# dataset.drop(columns=["Relevance"], inplace=True)
dataset.dropna(inplace=True)
dataset.drop_duplicates(inplace=True)
dataset.info
```

```python
# Reformatting dataframe data to new dataframe
final_data = pd.DataFrame(columns=["Name", "Line"])
for i, row in dataset.iterrows():
  if RomanUrdu:
    user = row["Intent (Roman Urdu)"] 
    response = row["Response (Roman Urdu)"]
  else:
    user = row["Intent"] 
    response = row["Response"]

  # Remove all special chars other than punctuation
  clean_pattern = re.compile('[^a-zA-Z0-9\.,;:?!\'\" ]+')
  user = clean_pattern.sub('', user)
  response = clean_pattern.sub('', response)
  # print(response)
  # print(user)

  final_data = final_data.append({"Name": "User", "Line": user}, ignore_index=True) 
  final_data = final_data.append({"Name": "Mai", "Line": response}, ignore_index=True)
```

```python
contexted = []
n = 7 # Context length 

# Creating new row with contexts and adding to dataframe
for i in range(n, len(final_data['Line'])):
  row = []
  prev = i - 1 - n # We additionally substract 1, so row will contain current response and 7 previous responses  
  for j in range(i, prev, -1):
    row.append(final_data['Line'][j].strip())
  contexted.append(row)  
```

```python
len(contexted)
```

```python
# Making list of columns according to context len
columns = ['response', 'context'] 
columns = columns + ['context/'+str(i) for i in range(n-1)]
columns
```

```python
df = pd.DataFrame.from_records(contexted, columns=columns) # Make dataframe
df.head(5)
```
